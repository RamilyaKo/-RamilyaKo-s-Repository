Файл df3.csv находится в архиве с заданием

```python
# Запустите эту ячейку
import pandas as pd
%matplotlib inline

url = 'https://drive.google.com/file/d/1qyOiTgEQAMPgHZaSiy36dnYOUcjtgzTw/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df3 = pd.read_csv(path)
print(len(df3))
print(df3.head())
```

Эти данные представляют собой заводские производственные показатели и сообщенное количество дефектов в определенные дни недели.

### 1. Воссоздайте этот график разброса «произведенных» и «дефектных»  деталей. Обратите внимание на цвет и размер точек. Также обратите внимание на размер картинки. Сможете ли растянуть его аналогичным образом?

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8.5, 2.5), dpi=95)
plt.scatter(df3['produced'],df3['defective'], color='red',alpha=0.5, s=7)
plt.title('График разброса «произведенных» и «дефектных» деталей')
plt.xlabel('produced')
plt.ylabel('defective')
plt.show()
```
